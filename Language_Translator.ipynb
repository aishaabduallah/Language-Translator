```python
import pandas as pd
```

```python

# قراءة ملف الإكسل
df = pd.read_excel('/content/قاموس.xlsx')
```

```python
df.head()
```

```python
# تنظيف البيانات (تحويل الإنجليزي لحروف صغيرة)
df['English'] = df['English'].str.lower().str.strip()
df['العربية'] = df['العربية'].str.strip()

print(" Language Translator")
print("1: English → Arabic")
print("2: Arabic → English")

choice = input("Choose option (1 or 2): ").strip()
```

```python

#  ترجمة من إنجليزي إلى عربي
if choice == "1":
    word = input("Enter English word: ").lower().strip()

    result = df[df['English'] == word]

    if not result.empty:
        print(" Translation:", result['العربية'].values[0])
    else:
        print(" Word not found")

#  ترجمة من عربي إلى إنجليزي
elif choice == "2":
    word = input("Enter Arabic word: ").strip()

    result = df[df['العربية'] == word]

    if not result.empty:
        print(" Translation:", result['English'].values[0])
    else:
        print(" Word not found")

#  اختيار خاطئ
else:
    print(" Invalid choice")
```
